## <center>Algoritmo de detección de fallas en generadores fotovoltaicos</center>

El presente trabajo tiene el objetivo de elaborar un algoritmo para la detección automática de fallas y el diagnóstico de posibles fallas que ocurren en una planta fotovoltaica (PFV) utilizando métodos estadísticos. El enfoque se ha validado utilizando datos experimentales de parámetros climáticos y eléctricos basados en una planta de 1,98 kWp instalada en la empresa Luz del Sur en la localidad San Juan de Ihauanco, perteneciente a la provincia de San Luis de Cañete. Existen pocas instancias de herramientas estadísticas que se despliegan en el análisis de los datos medidos de PV. El objetivo principal de este documento es, por lo tanto, crear un sistema capaz de simular los rendimientos teóricos de los sistemas fotovoltaicos y permitir el análisis estadístico de los datos medidos de la energía fotovoltaica. El algoritmo de detección de fallas compara la potencia de salida teórica y la medida utilizando una prueba t estadística. Para determinar la ubicación de la falla, se supervisa la relación entre la potencia y el voltaje teóricos y medidos. Los resultados obtenidos indican que el algoritmo de detección de fallas puede detectar.

## <center> Cálculo teórico de la potencia y voltaje</center>

La siguiente ecuación corresponde al modelamiento teórico de una celda fotovoltaica, el cual se optiene del circuito RC, este modelo fue utilizado ampliamente en trabajos anteriores. 

\begin{equation}
I_{a} = I_{ph}-I_{0}\left( e^{\frac{V_{a} + R_{s}I_{a}}{V_{t}N_{s}-1}} \right)-\frac{V_{a}+R_{s}I_{a}}{R_{sh}}
\label{Ia}
\tag{0}
\end{equation}

La ecuación \ref{Ia} es el modelo obtenido del circuito RC, que representa la dependencia no lineal entre la corriente $(I_{a})$ y el voltaje $(V_{a})$ de salida de una celda fotovoltaica. Sin embargo existen valores desconocidos de los parámetros tales como $R_{s}$, $R_{sh}$ y $V_{t}$, los cuales serán calculados con las siguientes ecuaciones:

## <center> Sistema de ecuaciones no lineales</center>

El conjunto de ecuaciones fue obtenido de la referencia [5]. Donde los parámetros de entrada constantes son: $I_{sc}$, $V_{oc}$, $V_{mpp}$, $I_{mpp}$, $n_{s}$ y $n_{p}$.

\begin{equation}
I_{mpp} = I_{sc} - \frac{V_{mpp} + I_{mpp}R_{s} - I_{sc}R_{s}}{R_{sh}} - \left( I_{sc}-\frac{V_{oc}-I_{sc}R_{s}}{R_{sh}}\right)e^{\frac{V_{mpp}+I_{mpp}R_{s}-V_{oc}}{n_{s}V_{t}}}
\tag{1}
\label{golden-mean}
\end{equation}

\begin{equation}
I_{mpp} + V_{mpp}\frac{\frac{(I_{sc}R_{sh}-V_{oc}+I_{sc}R_{s})e^{\frac{V_{mpp}+I_{mpp}R_{s}-V_{oc}}{n_{s}V_{t}}}}{n_{s}V_{t}R_{sh}} - \frac{1}{R_{sh}}} {1+ \frac{(I_{sc}R_{sh}-V_{oc}+I_{sc}R_{s})e^{\frac{V_{mpp}+I_{mpp}R_{s}-V_{oc}}{n_{s}V_{t}}}}{n_{s}V_{t}R_{sh}} + \frac{R_{s}}{R_{sh}}} = 0
\tag{2}
\label{mean} 
\end{equation}


\begin{equation}
\frac{-1}{R_{sh}} = \frac{\frac{-(I_{sc}R_{sh}-V_{oc}+I_{sc}R_{s})e^{\frac{I_{sc}R_{s}-V_{oc}}{n_{s}V_{t}}}}{n_{s}V_{t}R_{sh}} - \frac{1}{R_{sh}}} {1+\frac{(I_{sc}R_{sh}-V_{oc}+I_{sc}R_{s})e^{\frac{I_{sc}R_{s}-V_{oc}}{n_{s}V_{t}}}}{n_{s}V_{t}R_{sh}}+ \frac{R_{s}}{R_{sh}}}
\tag{3}
\end{equation}

\begin{equation}
I_{0} = \left( I_{sc} - \frac{V_{oc} - I_{sc}*R_{s}}{R_{sh}} \right) * e^{-\frac{V_{oc}}{n_{s}V_{t}}}
\label{Io}
\tag{4}
\end{equation}
